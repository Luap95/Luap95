```python
# prompt: mescle todas as planilhas que estão em disco e acresente uma coluna com o nome do arquivo, deve ser extraido somente o texto antes do -

import pandas as pd
import os
import re

def merge_excel_files(directory):
    """
    Merges all Excel files in a directory into a single DataFrame,
    adding a column with the filename (extracted text before '-').

    Args:
        directory: The path to the directory containing the Excel files.

    Returns:
        A pandas DataFrame containing the merged data, or None if no files
        are found or an error occurs.
    """

    all_data = []
    for filename in os.listdir(directory):
        if filename.endswith(('.xls', '.xlsx')):
            filepath = os.path.join(directory, filename)
            try:
                df = pd.read_excel(filepath, usecols=[0,1,2,3,4,5,6])

                # Extract filename before '-' using regex
                match = re.match(r"(.+?)-", filename)
                if match:
                    file_prefix = match.group(1)
                else:
                    file_prefix = filename  # Use the whole filename if no '-' is found

                df['Clinica'] = file_prefix
                all_data.append(df)

            except Exception as e:
                print(f"Error reading file {filename}: {e}")
                return None # or handle error differently

    if not all_data:
        print("No Excel files found in the specified directory.")
        return None

    merged_df = pd.concat(all_data, ignore_index=True)
    return merged_df


# Example usage: replace with your directory
directory_path = '/content/'
merged_data = merge_excel_files(directory_path)


```
